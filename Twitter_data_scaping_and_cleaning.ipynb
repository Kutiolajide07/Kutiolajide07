## Objectives:
1. Scrape/download twitter data in three languages (i.e. English, Russian, and Ukrainian). This will be done as:
    a. Using English hashtag ("Russian-UkrainianWar", "UkrainianRussianWar", "RussianWar", and “RussianUkrainianWar”). The English hashtag is not limited to English tweets, meaning that, it will also get tweets in the Ukrainian and Russian languages.
    b. The English hashtags will also be translated to Russian and Ukrainian languages and will be used to download Twitter data.
    c. We will combine Ukrainian tweets from 1a and 1b, and also combine Russian data from 1a and 1b. In the end, we have three datasets: English tweets, Russian tweets, and Ukrainian tweets.
2. Sentiment analysis: sentiment analysis will be done on the three datasets separately and the results will be compared in terms of sentiments gotten from each language. Note: since the student doesn’t understand Russian and Ukrainian languages, all tweets in Russian and Ukrainian languages will be translated to the English language before analysis.
3. Topic modeling: LDA will be used to identify prominent topics in the tweets from the three separate languages.
4. Cluster analysis; tweets from the three languages will be grouped into clusters separately. Here, our aim is to see whether positive, negative, and neutral tweets are grouped into separate clusters.

```python
df2 = pd.DataFrame()
for csv in csv2:
    df2 = df2.append(pd.read_csv(csv, low_memory=True), ignore_index=True)
```

```python
df2.shape
```

```python
#concatenate json1 and json2 data
df3 = pd.DataFrame()
for json in json1:
    df3 = df3.append(pd.read_json(json, lines=True), ignore_index=True)
```

```python
df4 = pd.DataFrame()
for json in json2:
    df4 = df4.append(pd.read_json(json, lines=True), ignore_index=True)
```

```python
#concatenate all the dataframes
full_df = pd.concat([df, df2, df3, df4], ignore_index=True)
full_df.shape
```

```python
#select the required languages only
new_df = full_df[full_df.language.isin(['en', 'ru', 'uk'])]
new_df.shape
```
